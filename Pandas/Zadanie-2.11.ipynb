# E. jaki jest najlepszy film wszechczasów, (najlepszy, czyli ma najwyższą średnią ocenę) - to zadanie możesz rozwiązać wykonując złączenie (join) zbioru movies i ratings

https://pandas.pydata.org/pandas-docs/stable/merging.html

```python
movies.head()
```

```python
ratings.head()
```

```python
pd.merge(movies, ratings).head()
```

```python
srednieOceny = ratings.groupby("MovieID")["Rating"].mean().reset_index()
pd.merge(srednieOceny, movies).sort_values(by="Rating", ascending=False)
```

# F. wykonaj poprzedni punkt, odrzucając wcześniej filmy które nie uzyskały wystarczająco dużo głosów (np 100)

```python
srednieOceny = ratings.groupby("MovieID").agg({"Rating":["mean", "size"]}) # nie tylko srednia, ale tez licznosc
```

```python
srednieOceny.columns
```

```python
srednieOceny100 = srednieOceny[srednieOceny[("Rating", "size")] > 100]
srednieOceny100.columns = ["mean", "size"]
pd.merge(srednieOceny100, movies, left_index=True, right_on="MovieID")\
                                .sort_values(by="mean", ascending = False)
```

# G. jaki jest najlepszy film według kobiet i według mężczyzn

```python
ratings_users = pd.merge(ratings, users)
```

```python
srednieOcenyPlcie = ratings_users.groupby(["MovieID", "Gender"]).agg({"Rating" : ["mean", "size"]})
srednieOcenyPlcie.columns = ["mean", "size"]
srednieOcenyPlcie = srednieOcenyPlcie.reset_index()
srednieOcenyPlcie100 = srednieOcenyPlcie[srednieOcenyPlcie["size"] > 100]
pd.merge(srednieOcenyPlcie100, movies).sort_values(by="mean", ascending=False)\
                                                    .drop_duplicates(subset="Gender")
```

# H. jaki jest średni rok oglądanego filmu w poszczególnych grupach wiekowych

```python
# polaczyc wszystkie 3 tabele (merge)
# policzyc sredni rok w podgrupach wiekowych

wszystko = pd.merge(pd.merge(movies, ratings), users)
```

```python
# tutaj mam year jako int
wszystko.groupby("Age")["Year"].mean()
```

# I. jakie trzy gatunki filmowe są najczęściej oglądane przez kobiety i mężczyzn

```python
# podpunkt D, ale na tabeli wszystko, w rozbiciu na płcie
mezczyzni = wszystko[wszystko.Gender == "M"]
kobiety = wszystko[wszystko.Gender == "F"]
```

```python
mezczyzni.Genres.str.split("|", expand=True).stack().reset_index(drop=True).value_counts()
```

```python
kobiety.Genres.str.split("|", expand=True).stack().reset_index(drop=True).value_counts()
```
