## Tabela przestawna

```python
pd.pivot_table(tips[["smoker", "total_bill", "tip"]],
               index=["smoker"],
               aggfunc="mean"
              )
```

```python
pd.pivot_table(tips[["smoker", "sex", "total_bill", "tip"]],
               index=["smoker"],
               columns=["sex"],
               aggfunc="mean"
              )
```

```python
pd.pivot_table(tips[["smoker", "sex", "total_bill", "tip"]],
               index=["smoker", "sex"],
               aggfunc="mean"
              )
```

```python
pd.pivot_table(tips[["smoker", "sex", "total_bill"]],
               index=["sex"],
               columns="smoker",
               aggfunc="min"
              )
```

```python
pd.pivot_table(tips[["smoker", "sex", "day", "total_bill"]],
               index=["sex", "day"],
               columns="smoker",
               aggfunc="median"
              )
```
