### E. jaki jest najlepszy film wszechczasów, (najlepszy, czyli ma najwyższą średnią ocenę) - to zadanie możesz rozwiązać wykonując złączenie (join) zbioru movies i ratings

```python
pd.merge(movies, ratings).head()
```

```python
srednieoceny = ratings.groupby([],['MovieID'])['Rating'].mean().reset_index()
pd.merge(srednieoceny, movies).sort_values(by='Rating',ascending=False).head()
```

### F. wykonaj poprzedni punkt, odrzucając wcześniej filmy które nie uzyskały wystarczająco dużo głosów (np 100)

```python
srednieoceny = ratings.groupby('MovieID').agg({'Rating':['mean','size']})
```

```python
srednieoceny100 = srednieoceny[srednieoceny[('Rating','size')] > 100]
srednieoceny100.columns = ['mean','size']
pd.merge(srednieoceny100,movies, left_index=True, right_on='MovieID').sort_values(by='mean', ascending = False)
```

### G. jaki jest najlepszy film według kobiet i według mężczyzn

```python
uzytkownicy = pd.merge(ratings,users)
uzytkownicy.head()
```

```python
uzytkownicyplec = uzytkownicy.groupby(['MovieID','Gender']).agg({'Rating':['mean','size']})
uzytkownicyplec.columns = ['mean','size']
uzytkownicyplec = uzytkownicyplec.reset_index()
uzytkownicyplec = uzytkownicyplec[uzytkownicyplec['size'] > 100]
uzytkownicyplec.head()
```

```python
pd.merge(uzytkownicyplec,movies).sort_values(by='mean', ascending = False).drop_duplicates(subset='Gender')
```

### H. jaki jest średni rok oglądanego filmu w poszczególnych grupach wiekowych

```python
wszystko = pd.merge(pd.merge(ratings, users),movies)
pd.merge(pd.merge(ratings, users),movies).groupby('Age')['Year'].mean()
```

```python
mezczyzni = wszystko[wszystko['Gender']=='M']
kobiety = wszystko[wszystko['Gender']=='F']
```
