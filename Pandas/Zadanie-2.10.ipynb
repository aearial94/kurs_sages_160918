## groupby()

```python
planes.groupby(["engine", "type"]).size().reset_index()\
                            .rename({0:"COUNT(*)"}, axis=1)[["COUNT(*)","engine", "type"]]
```

```python
planes.groupby(["engine", "type"]).size().reset_index()\
                            .rename({0:"COUNT(*)"}, axis=1).loc[:,::-1]
```

## tabela kontyngencji

```python
pd.crosstab(planes.engine, planes.type)
```

```python
wynik = pd.crosstab(planes.engine, planes.type).stack()
wynik
```

```python
pd.crosstab(planes.engine, planes.type, margins=True)
```

```python
wynik = pd.crosstab(planes.engine, planes.type).unstack()
wynik = wynik.reset_index().rename({0 : "COUNT(*)"},axis=1)
wynik[wynik["COUNT(*)"] > 0].reset_index(drop=True)
```

## tabela przestawna

```python
# te same trudnosci co przy crosstab, ale tutaj da sie je latwo ominac
wynik = pd.pivot_table(planes[["engine", "type"]],
               index=["engine"],
               columns = "type",
               aggfunc="size"
              ).stack().reset_index().rename({0 : "COUNT(*)"}, axis=1)
wynik["COUNT(*)"] = wynik["COUNT(*)"].astype(int)
wynik
```

```python
pd.pivot_table(planes[["engine", "type"]],
               index=["engine", "type"],
               aggfunc="size"
              ).reset_index()
```

## groupby()

```python
# znaleźć funkcję, która w jednym ruchu policzy nam 3 funkcje agregujące: min, mean, max
planes.groupby(["engine", "manufacturer"])["year"].agg(["min", "mean", "max"])
```

```python
wynik = planes.groupby(["engine", "manufacturer"], as_index=False)\
                            .agg({"year" : ["min", "mean", "max"]})
wynik["engine"]
wynik[("year", "min")]
wynik[[("year", "min"), ("year", "max")]]
wynik[[("engine", ""), ("manufacturer", ""), ("year", "min"), ("year", "mean"), ("year", "max")]]
```

```python
wynik = planes.groupby(["engine", "manufacturer"]).describe()
```

```python
wynik = wynik[[("year", "min"), ("year", "mean"), ("year", "max")]].reset_index()
wynik.columns = ["engine", "manufacturer", "MIN(year)", "AVG(year)", "MAX(year)"]
wynik
```

## pivot table

```python
pd.pivot_table(planes[["engine", "year", "manufacturer"]],
               index=["engine", "manufacturer"],
               aggfunc=["min", "mean", "max"]
              ).reset_index()
```
